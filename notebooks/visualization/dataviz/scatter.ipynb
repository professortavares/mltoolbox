# Notebook responsible for displaying a scatter plot from randomly generated data

```python
import numpy as np
import pandas as pd

from mltoolbox.visualization import DataViz
```

## Generates a set of 1000 random numbers

```python
mu, sigma = 0, 1 # mean and standard deviation
a = np.random.normal(mu, sigma, 1000)
# linear relation between a and b
b = 5*a + 3

d = {
    'a':a,
    'b':b
}

df = pd.DataFrame(d)
```

```python
df
```

## Creates an instance of DataViz and plots the scatter plot

```python
dv = DataViz()
dv.plot_scatterplot(df=df, x='a', y='b')
```
